```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from wordcloud import STOPWORDS

 
font_path = 'c:\\windows\\fonts\\malgun.ttf'
stopwords = set(STOPWORDS)
stopwords.add('china')
stopwords.add('new')
stopwords.add('now')
stopwords.add('s')
stopwords.add('t')
stopwords.add('monday')
stopwords.add('tuesday')
stopwords.add('wednesday')
stopwords.add('thursday')
stopwords.add('friday')
stopwords.add('saturday')
stopwords.add('sunday')
stopwords.add('year')
stopwords.add('briefing')

wordcloud = WordCloud(
    font_path = font_path,
    stopwords=stopwords,
    width = 800,
    height = 800,
    background_color='white'
)

text=open('C:/Users/penpi/Desktop/nytimes-korea culture.prn').read()
wordcloud = wordcloud.generate(text)

fig = plt.figure(figsize=(12,12))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
fig.savefig('wordcloud_without_axisoff.png')
```

```python
import numpy as np
from PIL import Image
from wordcloud import ImageColorGenerator

ny_coloring=np.array(Image.open('C:/Users/penpi/Desktop/map.png'))
image_colors=ImageColorGenerator(ny_coloring)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

ny_mask = np.array(Image.open('C:/Users/penpi/Desktop/map.png'))
wordcloud = WordCloud(stopwords=stopwords, background_color='white', mask=ny_mask, max_font_size=200).generate(text)

plt.figure(figsize=(12,12))
plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation='bilinear')
plt.axis('off')
plt.show()
```
